-  1. Layout Direction
Flexbox: One-dimensional layout model. You can lay out items in a single row OR a single column.

Grid: Two-dimensional layout model. You can lay out items in both rows AND columns.

2. Content vs Container Based
Flexbox: More content-oriented. It adjusts to the size of the content inside.

Grid: More container-oriented. You define the grid structure (rows and columns) ahead of time.

3. Item Placement
Flexbox: Items are laid out in a linear flow and aligned using properties like justify-content, align-items, and flex.

Grid: Allows explicit placement of items using grid-row, grid-column, or even grid-template-areas.

 4. Complex Layouts
Flexbox: Best for simpler layouts like navigation bars, toolbars, buttons in a row, or vertical stacks.

Grid: Ideal for complex page layouts like web templates, dashboards, and galleries.

 5. Overlapping Elements
Flexbox: Does not support overlapping elements.

Grid: Supports overlapping elements with z-index and grid line control.

Q2- 2. Explain the role of the following key properties in the Flexbox layout ?
1 justify-content
2 align-item
3 gap
4 flex-direction
5 flex-wrap

1. justify-content
This property aligns the flex items along the main axis (which is horizontal by default).

Purpose: To control the horizontal spacing and alignment of items inside the container.

Common values:

flex-start: Aligns items to the start of the main axis.

center: Centers items along the main axis.

flex-end: Aligns items to the end of the main axis.

space-between: Evenly distributes items with space between them.

space-around: Adds equal space around each item.

space-evenly: Distributes space evenly between and around items.

2. align-items
This property aligns flex items along the cross axis (vertical by default).
Purpose: To control vertical alignment of items inside the container.
Common values:

stretch (default): Items stretch to fill the container’s height.

flex-start: Aligns items to the top of the container.

center: Centers items vertically.

flex-end: Aligns items to the bottom.

baseline: Aligns items by their text baseline.

3. gap
The gap property defines the spacing between flex items.

Purpose: To set horizontal and/or vertical space between items without using margins.

Example:
display: flex;
gap: 10px;
This adds 10px of space between each item.

4. flex-direction
This property defines the direction of the main axis along which the items are laid out.

Purpose: To control whether items are placed in a row or a column.

Values:

row (default): Items are placed from left to right.

row-reverse: Items are placed from right to left.

column: Items are placed from top to bottom.

column-reverse: Items are placed from bottom to top.

5. flex-wrap
This property controls whether the items should wrap onto multiple lines.

Purpose: To handle overflow when there are too many items for one line.

Values:

nowrap (default): All items stay on one line, may overflow.

wrap: Items will wrap onto the next line if needed.

wrap-reverse: Items wrap to the next line in the opposite direction.
